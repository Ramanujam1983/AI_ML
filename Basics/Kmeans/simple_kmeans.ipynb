```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd

# Step 1: Data
data = np.array([
    [15, 39],
    [16, 81],
    [17, 6],
    [18, 77],
    [19, 40],
    [20, 90]
])
customers = ['A','B','C','D','E','F']

# Step 2: Run KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(data)

# Step 3: Show results
df = pd.DataFrame(data, columns=['Income','SpendingScore'])
df['Customer'] = customers
df['Cluster'] = kmeans.labels_
print(df)
print("\nCentroids:\n", kmeans.cluster_centers_)
```
